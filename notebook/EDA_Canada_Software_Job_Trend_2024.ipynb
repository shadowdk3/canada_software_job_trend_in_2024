### 4. Exploring Data ( Visualization )


```python
def top_10_avg_salary(col_name):
    grouped = df.groupby(col_name)

    # Calculate the average salary, maximum salary, and minimum salary for each job info
    avg_salary = grouped['Avg_Salary'].mean()
    avg_max_salary = grouped['Max_Salary'].mean()
    avg_min_salary = grouped['Min_Salary'].mean()

    # Merge the DataFrames together
    merged_data = pd.merge(avg_salary, avg_max_salary, on=col_name, suffixes=('_avg', '_max'))
    merged_data = pd.merge(merged_data, avg_min_salary, on=col_name)

    # Reset the index for sorting
    merged_data.reset_index(inplace=True)

    # Get the top 10 job info entries based on average salary
    top_10_avg_salary = merged_data.nlargest(10, 'Avg_Salary')

    top_10_avg_salary['Avg_Salary'].min()
    min_salary_row = top_10_avg_salary[top_10_avg_salary['Avg_Salary'] == top_10_avg_salary['Avg_Salary'].min()]

    for index, row in min_salary_row.iterrows():
        print(f"min salary {col_name}:", row[col_name])

    return top_10_avg_salary
```

```python
def plot_top_10_salary(top_10_avg_salary, col_name):
    plt.figure(figsize=(16, 8))

    plt.bar(top_10_avg_salary[col_name], top_10_avg_salary['Avg_Salary'], color='b', alpha=0.7, label='Average Salary')
    plt.bar(top_10_avg_salary[col_name], top_10_avg_salary['Max_Salary'], color='g', alpha=0.7, label='Max Salary')
    plt.bar(top_10_avg_salary[col_name], top_10_avg_salary['Min_Salary'], color='r', alpha=0.7, label='Min Salary')

    plt.xlabel(col_name)
    plt.ylabel('Salary')
    plt.title(f'Top 10 Salary {col_name}')
    plt.xticks(rotation=90)
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
def group_salary(col_name):
    grouped = df.groupby(col_name)

    avg_salary = grouped['Avg_Salary'].mean()
    avg_max_salary = grouped['Max_Salary'].mean()
    avg_min_salary = grouped['Min_Salary'].mean()

    merged_data = pd.merge(avg_salary, avg_max_salary, on=col_name, suffixes=('_avg', '_max'))
    merged_data = pd.merge(merged_data, avg_min_salary, on=col_name)

    return merged_data
```

```python
def plot_KDE(df, col_name):
    fig, axs = plt.subplots(1, 2, figsize=(15, 7))
    plt.subplot(121)
    sns.histplot(data=df,x='Avg_Salary',bins=30,kde=True,color='g')
    plt.subplot(122)
    sns.histplot(data=df,x='Avg_Salary',kde=True,hue=col_name)
    plt.show()
```
